```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
# reading the dataset
file_path=r"D:\HH CPBA\CarPrice_Assignment.csv"
df= pd.read_csv(file_path ,index_col=0)
df.head()
```

```python
df.info()
```

```python
#splitting the carname column into it's brand and model
df = df.join(df['CarName'].str.split(' ', 1, expand=True).rename(columns={0:'CarCompany', 1:'CarModel'}))
df=df.drop(['CarName'],axis=1)
df.head()
```

```python
#Remove the starting and leading whitespaces for string objects
str_cols = df.select_dtypes(['object'])
df[str_cols.columns] = str_cols.apply(lambda x: x.str.strip())
df.head()
```

```python
#Convert all string columns data to uppercase
df[str_cols.columns] = str_cols.apply(lambda x: x.str.upper())
df.head()
```

```python
#mapping categorical variables into numerical values
df['fueltype']=df['fueltype'].map({'GAS': 1, 'DIESEL': 0})
df['enginelocation']=df['enginelocation'].map({'FRONT': 1, 'REAR': 0})
df['aspiration']=df['aspiration'].map({'STD': 1, 'TURBO': 0})
df['doornumber']=df['doornumber'].map({'TWO': 1, 'FOUR': 0})
df['cylindernumber']=df['cylindernumber'].map({'TWO': 2, 'THREE': 3,'FOUR': 4,'FIVE': 5,'SIX':6,'EIGHT':8,'TWELVE':12})
```

```python
df.info()
```

```python
df['area_car'] = df['carlength'] * df['carwidth']
df['volume_car'] = df['carlength'] * df['carwidth'] * df['carheight']
df['weight_to_volume_ratio'] = df['curbweight'] / df['volume_car']
df['engsize_to_power_ratio'] = df['horsepower'] / df['enginesize']
df['avg_milage'] = (df['citympg']+df['highwaympg'])/2
df['weight_to_milage_ratio'] = (df['avg_milage']/df['curbweight'])
```

```python
symboling = pd.get_dummies(df['symboling'],drop_first=True,prefix='symboling')
#Adding the results to the master dataframe
df = pd.concat([df,symboling],axis=1)
df=df.drop(['symboling'],axis=1)
df.head()
```

```python
carbody = pd.get_dummies(df['carbody'],drop_first=True,prefix='carbody')
#Adding the results to the master dataframe
df = pd.concat([df,carbody],axis=1)
df=df.drop(['carbody'],axis=1)
df
```

```python
drivewheel = pd.get_dummies(df['drivewheel'],prefix='drivewheel')
#Adding the results to the master dataframe
df = pd.concat([df,drivewheel],axis=1)
df=df.drop(['drivewheel'],axis=1)
```

```python
enginetype = pd.get_dummies(df['enginetype'],prefix='enginetype')
#Adding the results to the master dataframe
df = pd.concat([df,enginetype],axis=1)
df=df.drop(['enginetype'],axis=1)
```

```python
fuelsystem = pd.get_dummies(df['fuelsystem'],prefix='fuelsystem')
#Adding the results to the master dataframe
df = pd.concat([df,fuelsystem],axis=1)
df=df.drop(['fuelsystem'],axis=1)
```
